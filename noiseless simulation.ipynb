```python
from vqe_algorithm import vqe_algorithm, exact_solution
from observables import *
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 5
result, history, circuit = vqe_algorithm(N, shots = 4096, maxiter = 200)
theta_opt = result.x

E_exact = exact_solution(N)
error = relative_error(E_exact, np.array(history))

# Plot convergence of the VQE
plt.figure(figsize=(7,4))
plt.semilogy(error)
plt.xlabel('cost-function evaluations')
plt.ylabel('relative error')
plt.title(f'VQE convergence, N = {N}')
plt.grid(True, which='both', alpha=0.3)
plt.show()
```

```python
N = 5
h_values = np.linspace(0.1, 2.0, 50)

data = np.zeros((3, len(h_values)), float)

for i in range(len(h_values)):
    h = h_values[i]
    
    result, _, circuit = vqe_algorithm(N,h=h, reps = 3, shots=4096)
    theta_opt = result.x

    data[0, i] = compute_expectation(theta_opt, circuit, H(N, h=h), shots=4096)
    data[1, i] = mean_magnetization(theta_opt, circuit, N)
    data[2, i] = mean_correlation(theta_opt, circuit, 1, N, shots=4096)

# ---- Energy plot ----
plt.figure()
plt.plot(h_values, data[0], marker='o', label="Energy")
plt.xlabel("Magnetic field h")
plt.ylabel("Energy ⟨H⟩")
plt.title(f"Ground State Energy vs Magnetic Field (N={N})")
plt.grid(True)
plt.legend()
plt.show()

# ---- Magnetization plot ----
plt.figure()
plt.plot(h_values, data[1], marker='s', color='orange', label="Magnetization")
plt.xlabel("Magnetic field h")
plt.ylabel("⟨M⟩")
plt.title(f"Mean Magnetization vs Magnetic Field (N={N})")
plt.grid(True)
plt.legend()
plt.show()

# ---- Correlation plot ----
plt.figure()
plt.plot(h_values, data[2], marker='^', color='green', label="Correlation")
plt.xlabel("Magnetic field h")
plt.ylabel("⟨Z_i Z_i+1⟩")
plt.title(f"Spin Correlation vs Magnetic Field (N={N})")
plt.grid(True)
plt.legend()
plt.show()
```
